```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.polynomial as polynomial
```

```python
from lagrange import Lagrange, GeneralizedLagrange, plot
```

```python
f = lambda x: np.sin(3 * np.pi * x)
```

```python
n = 11  # number of samples

x = polynomial.chebyshev.chebroots([0] * n + [1])
y = f(x)
```

```python
simple = Lagrange(x, y)
generalized = GeneralizedLagrange(x, y, phi=lambda x: np.sin(x))
```

```python
x_test = np.linspace(x.min(), x.max(), 10 * n)
y_test = f(x_test)
p_simple = simple(x_test)
p_generalized = generalized(x_test)
```

```python
plot(x_sample=x, y_sample=y, x_pred=x_test, y_pred=p_simple, y_real=y_test)
```

```python
plot(x_sample=x,
     y_sample=y,
     x_pred=x_test,
     y_pred=p_generalized,
     y_real=y_test)
```
