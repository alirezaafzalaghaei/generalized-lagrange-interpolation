```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.polynomial as polynomial
```

```python
from lagrange import Lagrange, plot
```

```python
f = lambda x: 1 / (1 + 25 * x**2)
```

```python
n = 11  # number of samples

x_equidistant = np.linspace(-1, 1, n)

# legendre and chebyshev roots are only defined on the interval [-1, 1]:
x_legroots = polynomial.legendre.legroots([0] * n + [1])
x_chebroots = polynomial.chebyshev.chebroots([0] * n + [1])

y_equidistant = f(x_equidistant)
y_legroots = f(x_legroots)
y_chebroots = f(x_chebroots)
```

```python
interpolate_equidistant = Lagrange(x_equidistant, y_equidistant)
interpolate_legroots = Lagrange(x_legroots, y_legroots)
interpolate_chebroots = Lagrange(x_chebroots, y_chebroots)
```

```python
x_test = np.linspace(-1, 1, 10 * n)
y_test = f(x_test)
p_equidistant = interpolate_equidistant(x_test)
p_legroots = interpolate_legroots(x_test)
p_chebroots = interpolate_chebroots(x_test)
```

```python
plot(x_sample=x_equidistant,
     y_sample=y_equidistant,
     x_pred=x_test,
     y_pred=p_equidistant,
     y_real=y_test)
```

```python
plot(x_sample=x_legroots,
     y_sample=y_legroots,
     x_pred=x_test,
     y_pred=p_legroots,
     y_real=y_test)
```

```python
plot(x_sample=x_chebroots,
     y_sample=y_chebroots,
     x_pred=x_test,
     y_pred=p_chebroots,
     y_real=y_test)
```

```python
plt.plot(x_test, y_test, label='Exact')
plt.plot(x_test, p_equidistant, label='Equidistant')
plt.plot(x_test, p_legroots, label='Legendre')
plt.plot(x_test, p_chebroots, label='Chebyshev')
plt.legend()
```
